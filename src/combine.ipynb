```python
# Import the os and time modules
import os
import time

# Define a function to get all the Javascript and JSX files in a given directory and its subdirectories
def get_js_files(dir):
    # Initialize an empty list to store the file paths
    js_files = []
    # Loop through the directory and its subdirectories
    for root, dirs, files in os.walk(dir):
        # Loop through the files
        for file in files:
            # Check if the file has a .js or .jsx extension
            if file.endswith(".js") or file.endswith(".jsx"):
                # Append the full file path to the list
                js_files.append(os.path.join(root, file))
    # Return the list of Javascript and JSX files
    return js_files

# Define a function to filter the files by a keyword
def filter_files(files, keyword):
    # Initialize an empty list to store the filtered files
    filtered_files = []
    # Split the keyword by commas and strip the whitespace
    keywords = [k.strip() for k in keyword.split(",")]
    # Loop through the files
    for file in files:
        # Check if any of the keywords is in the file name
        if any(k in file for k in keywords):
            # Append the file to the filtered list
            filtered_files.append(file)
    # Return the filtered list
    return filtered_files

# Define a function to filter the files by a keyword
def filter_files(files, keyword):
    # Initialize an empty list to store the filtered files
    filtered_files = []
    # Split the keyword by commas and strip the whitespace
    keywords = [k.strip() for k in keyword.split(",")]
    # Loop through the files
    for file in files:
        # Convert the file name and the keyword to lowercase
        file_lower = file.lower()
        keywords_lower = [k.lower() for k in keywords]
        # Check if any of the keywords is in the file name
        if any(k in file_lower for k in keywords_lower):
            # Append the file to the filtered list
            filtered_files.append(file)
    # Return the filtered list
    return filtered_files


# Define a function to write the combined content to a text file
def write_to_file(content, file_name):
    # Open the file in write mode
    with open(file_name, "w") as f:
        # Write the content to the file
        f.write(content)

# Define a function to concatenate the contents of the filtered files
def combine_files(files):
    combined_content = ""
    for file in files:
        with open(file, 'r') as f:
            combined_content += f.read() + "\n\n"
    return combined_content

# Get the current directory of the notebook
current_dir = os.path.abspath('')

# Get the list of Javascript and JSX files in the current directory and its subdirectories
js_files = get_js_files(current_dir)
# Ask the user to enter a keyword
keyword = input("Enter keywords separated by commas , to filter the files (or type 'all' to include all files): ")
# Check if the user entered 'all'
if keyword == "all":
    # Use the original list of files
    filtered_files = js_files
else:
    # Filter the files by the keyword
    filtered_files = filter_files(js_files, keyword)
# Combine the content of the filtered files
combined_content = combine_files(filtered_files)
# Call the function with the combined content and a file name of your choice
write_to_file(combined_content, "js_files.txt")
```
